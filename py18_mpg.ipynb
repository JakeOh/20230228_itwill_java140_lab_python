## 연속형 변수 탐색

## 카테고리 타입 변수 탐색

```python
# origin(자동차 생산 지역) 빈도수
mpg.origin.value_counts()
```

```python
sns.countplot(data=mpg, x='origin')
plt.show()
```

```python
# cylinders 빈도수
mpg.cylinders.value_counts()
```

```python
sns.countplot(data=mpg, x='cylinders')
plt.show()
```

```python
# model_year 빈도수
mpg.model_year.value_counts()
```

```python
sns.countplot(data=mpg, x='model_year')
plt.show()
```

# 연비에 영향을 미치는 변수

## mpg(연비) ~ displacement(배기량)

```python
sns.scatterplot(data=mpg, x='displacement', y='mpg')
plt.show()
```

```python
sns.scatterplot(data=mpg, x='displacement', y='mpg', hue='origin')
plt.show()
```

```python
variables = ['cylinders', 'horsepower', 'weight', 'acceleration']  # 독립변수

# 2x2 sublot들을 생성:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

# 각각의 subplot에 산점도 그리기:
for i in range(2):
    for j in range(2):
        idx = 2 * i + j
        sns.scatterplot(data=mpg, x=variables[idx], y='mpg', ax=ax[i, j])

plt.show()
```
