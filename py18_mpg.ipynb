mpg(Mile Per Gallon) dataset: 자동차 연비 데이터 셋

*  cylinders: 실린더 개수

*  weight: 자동차 무게
*  acceleration: 가속도.
*  model_year: 자동차 생산 연도
*  origin: 자동차 생산지
*  name: 자동차 이름
